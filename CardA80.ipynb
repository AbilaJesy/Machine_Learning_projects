# Preprosessing -->Saler

```python
df.corr()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()

df['Time']=sc.fit_transform(df['Time'].values.reshape(-1, 1))
df['Amount']=sc.fit_transform(df['Amount'].values.reshape(-1, 1))
```

# Splitting X and Y features

```python
X =df.drop("Class", axis=1)
Y = df["Class"]
```

# Dataset is imbalance so using over sampling Preprocessing techniques

```python
from imblearn.over_sampling import SMOTE
oversampler = SMOTE(random_state=2)
X_new , Y_new = oversampler.fit_resample(X, Y)
```

# visuvalizing balanced data

```python
sns.countplot(x=Y_new)
```

# splitting dataset train and test

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X_new,Y_new,test_size=0.2,random_state=42)
```

# Algorithm Model_selection

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=20)
```

```python
#fit the training dataset
rf.fit(x_train,y_train)
```

# Predict the training dataset

```python
x_train_predict=rf.predict(x_train)
```

# Mertrics---> error rate

```python
from sklearn.metrics import accuracy_score as acc,recall_score as rs,precision_score as ps,f1_score as f1,confusion_matrix as cm,roc_curve,auc

print('Accuracy_score\t:',acc(y_train,x_train_predict))
print('recall_score\t:',rs(y_train,x_train_predict))
print('precision_score\t:',ps(y_train,x_train_predict))
print('f1_score\t:',f1(y_train,x_train_predict))
print('Confusion matrix:\n\n',cm(y_train,x_train_predict))
```

```python
fpr, tpr, thresholds = roc_curve(y_train, x_train_predict)
roc_auc = auc(fpr, tpr)
# Plot the ROC curve
plt.plot(fpr, tpr, lw=1, label='ROC curve (AUC = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], '--', color='gray', lw=1)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc="lower right")
plt.show()
```

# Predict the testing dataset

```python
x_test_predict=rf.predict(x_test)
```

```python
print('Accuracy_score\t:',acc(y_test,x_test_predict))
print('recall_score\t:',rs(y_test,x_test_predict))
print('precision_score\t:',ps(y_test,x_test_predict))
print('f1_score\t:',f1(y_test,x_test_predict))
print('Confusion matrix:\n\n',cm(y_test,x_test_predict))
```

```python
fpr, tpr, thresholds = roc_curve(y_test,x_test_predict)
roc_auc = auc(fpr, tpr)
# Plot the ROC curve
plt.plot(fpr, tpr, lw=1, label='ROC curve (AUC = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], '--', color='gray', lw=1)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc="lower right")
plt.show()
```
